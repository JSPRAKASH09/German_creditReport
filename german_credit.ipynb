```python
def missing_val_input(dataset,col):
    dataset.loc[dataset[col].isnull(),col]=dataset[col].median()
    print("missung values for {} done".format(col))
    return
```

```python
def missing_val_input_cat(dataset,col):
    dataset.loc[dataset[col].isnull(),col]=dataset[col].mode()[0]
    print("missung values for {} done".format(col))
    return


```
